# Matplotlib

Matplotlib es una librería escrita en python, esta enfocada en la visualización de datos por medio de gráficos 2D tales como:
- Barras
- Líneas
- Pie
- Dispersión

Entre muchos otros.

Esta herramienta se ha convertido en el estándar de creación de gráficos, ideal a la hora de realizar análisis exploratorio de datos gracias a su integración con las librerías de `numpy` y `pandas`. Además es bastante versátil pues permite personalizar los gráficos de forma avanzada.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 3.1 Gráfico de líneas

Permite mostrar tendencias y cambios en los datos a lo largo del tiempo.

**Ejemplo**: Analizar las ventas de un determinado producto a lo largo del tiempo, de esta manera identificamos como suben o bajan las ventas en un determinado periodo.

```python
np.random.seed(42)
# Datos
months = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
sales = np.random.randint(50, 200, 6)
```

```python
# Configuración del gráfico
plt.figure(figsize=(10,8))
plt.plot(
  months,
  sales,
  marker="o",
  color="blue"
)

plt.title("Ejemplo de ventas mensuales")
plt.xlabel("Meses")
plt.ylabel("Ventas en miles de unidades")
plt.show()
```

## 3.2 Gráfico de dispersión

Muestra la relación entre dos variables, permite encontrar patrones, agrupaciones o puntos atípicos en los datos.

**Ejemplo**: Analizar la relación entre las horas de estudio y las calificaciones obtenidas por los estudiantes; de esta forma podemos determinar si hay una correlación entre las dos variables (a más horas de estudio -> más alta la calificación).

```python
np.random.seed(42)
# Relación entre horas de esestudio y calidificaciones
hours = [0,1,2,3,4,5,6,7,8]
exam = np.random.randint(50,100, 9)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(
  x=hours,
  y=exam,
  marker="o"
)
plt.title("Ejemplo: Relación entre horas de estudio y calificacines")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación obtenida")
plt.show()
```

## 3.3 Gráfico de barras

Permite representar datos categóricos de forma visual y más comprensible. 

Éste tipo de gráfico compara cantidades (datos discretos) entre diferentes categorias.

**Ejemplo**: Comparar la venta de diferentes productos en un mes. <br>
**Aplicación**: Es útil para ver qué producto tiene más ventas y como es su desempeño con otros.

```python
np.random.seed(42)
# Datos
products = ["Producto A","Producto B","Producto C"]
sales = np.random.randint(20, 150, 3)
```

```python
# Creación del gráfico de barras
plt.figure(figsize=(10,8))
plt.bar(
  x=products,
  height=sales,
  label="Ventas mensuales",
  color="skyblue"
)

plt.title("Ventas de productos en un mes")
plt.xlabel("Productos")
plt.ylabel("Ventas")
plt.legend()

# Anotación para anotar un punto ezpecifico:
# El producto con mas ventas.
max_sales = sales.max()
max_sales_idx = np.where(sales == max_sales)[0][0]
max_prod = products[max_sales_idx]

plt.annotate(
  "Máximo de ventas", 
  xy=(max_prod, max_sales),
  arrowprops=dict(facecolor="red", shrink=0.05)  
)

plt.show()
```

## 3.4 Gráfico de pastel (pie)

Un gráfico de pie nos muestra la proporción de cada categoría en relación a un total, donde cada pedazo representa una categoría y su tamaño representa el valor de la proporción correspondiente.

⚠️ Utilizar solo cuando:
- Hay una diferencia clara en las proporciones.
- Son pocas categorías.


**Ejemplo**: Participación de diferentes productos en el mercado.

```python
# Datos
np.random.seed(42)
products = ["Producto A","Producto B","Producto C"]
market_share = np.random.randint(20, 150, 3)
```

```python
plt.figure(figsize=(10,8))
plt.pie(
  x=market_share,
  labels=products,
  autopct="%1.2f%%",
  startangle=140, 
  colors=["gold", "lightcoral", "orange"]
)

plt.title("Participación de mercado por producto")
plt.axis("equal");
```

## 3.5 Gráficos de distribución

Dentro de esta categoría de gráficos encontramos a los histogramas y boxplot, son herramientas visuales utilizadas para resumir  la distribución y dispersión de un conjunto de datos. 


Altamente utilizados en el análisis exploratorio de datos, ya que permiten visualizar como se distribuyen los datos y cómo es su variabilidad para detectar datos atípicos (anómalos).

### 3.5.1 Histograma

- Similar al gráfico de barras.
- Muestra la distribución de datos continuos.
- Visualiza la frecuencia de ocurrencia de diferentes rangos de valores.

**Ejemplo**: Distribución de las alturas de un grupo de personas.

```python
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=200)
print(data[:5])
```

📌 **Nota**: La función `normal` crea un array con datos que siguen una distribución normal, donde:
- `loc`: Valor de la media.
- `scale`: Valor de la desviación estándar.
- `size`: Cantidad de datos que va a generar.

```python
# Creamos un histograma personalizado:
plt.hist(
  x=data,
  color="salmon",
  edgecolor="black",
  alpha=0.6
)

plt.title("Distribución de alturas")
plt.xlabel("Altura (cm)")
plt.ylabel("Frecuencia");
```

Cada barra representa un intervalo de frecuencias que por default agrupa 10 rangos de valores, matplotlib nos permite modificar este intervalo mediante el parámetro `bins`.

```python
# Creamos un histograma personalizado:
bins = 20
plt.hist(
  x=data,
  color="salmon",
  edgecolor="black",
  alpha=0.6,
  bins=bins
)

plt.title(f"Distribución de alturas (bins: {bins})")
plt.xlabel("Altura (cm)")
plt.ylabel("Frecuencia");
```

### 3.5.2 Boxplot

- Muestra la distribución de datos mediante cuartiles.
- Útil para identificar valores atípicos, dispersión y simetría en los datos.

**Ejemplo**: Distribución de edades de diferentes grupos de personas.

```python
np.random.seed(42)
ages = [
  np.random.normal(loc=30, scale=5, size=100),
  np.random.normal(loc=40, scale=5, size=100),
  np.random.normal(loc=60, scale=5, size=100)
]
print(ages[0][:5], "\n", ages[1][:5], "\n", ages[2][:5])
```

```python
plt.boxplot(
  x=ages,
  patch_artist=True,
  notch=True,
  vert=False,
  labels=["Grupo 1", "Grupo 2", "Grupo 3"]
)

plt.title("Distribución de edades de 3 grupos")
plt.xlabel("Edades (años)")
plt.ylabel("Grupos")
```
